## Series 
    A pandas 1D labeled array that can hold any data type. Think of it like a single column in a spreadshit

```python
import pandas as pd

# we created python list with 3 values
data = [100, 102, 104, 106, 108, 110]
# let's convert to series object
series = pd.Series(data)
print(series)

# we can customize indexes as well
indexes = ["a", "b", "c", "d", "e", "f"]
series = pd.Series(data, index=indexes)
print(series)
# we can access values with indexes
print(series.loc["a"])


# to update value we can use loc as well
series.loc["a"] = 200
print(series)

# we can also access our values with position
print(series.iloc[0])

# to update value with position we can use iloc as well
series.iloc[5] = 300
print(series.iloc[5])

print(series)
```

```python
# lets filter by value
filtered_series = series[series > 105]
print(filtered_series)
```

```python
# dictionary can be converted to series as well
# dictionary made of key and value pairs, key will be index and value will be value of series
calories = {"Day 1" : 1750, "Day 2": 2100, "Day3": 1700}
series = pd.Series(calories)
print(series)

# we can access values with loc as well
print(series.loc["Day 1"])
# we can update values with loc as well
# we can update like this as well : series.loc["Day 1"] = 2000
# or like this:
series.loc["Day 1"] +=500 

print(series.loc["Day 1"])

print(series)

# filtering by value
filtered = series[series >= 2000]
print(filtered)
```

## DataFrame 
 A tabular data structure with rows AND columns. (2D) 
 Similar to an Excel spreadsheet


```python
data = { "Name": ["Spongebob", "Patrick", "Squidward"], 
         "Age": [30, 35, 50]
}
# we will convert this dictionary to dataframe

# we can customize indexes as well ( or we could use default indexes 0,1,2)
indexes = ["Employee 1", "Employee 2", "Employee 3"]
df = pd.DataFrame(data, index=indexes)
print(df)

# to select single row we can use loc
print(df.loc["Employee 1"])

# we can also select with indexes
print(df.iloc[0])

# to add a new column we can simply assign values to new column name
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)

# to add a new row we simply create a new data frame and concatenate it with existing dataframe
df_with_new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Scientist"},
                                 { "Name": "Mr. Krabs", "Age": 70, "Job": "Businessman"}], 
                               index = ["Employee 4", "Employee 5"])
df = pd.concat([df, df_with_new_row])
print(df)

```

## Importing

```python
# we are going to create dataframe using our csv data
# reading files from csv
df = pd.read_csv("weatherYVR.csv")
print(df)
```

## Selection


```python
# selection by column name
print(df["Year"])
# selecting multiple columns 
print(df[["Year","Month"]])

# selection by row ( by index)
print(df.loc[0])
# selection by position
print(df.iloc[0:2])
```

## Filtering
Keeping the rows that match the condition

```python
# let's say we want to keep only the december months data, that is month = 12
# we can create new dataframe with only december data
december_data = df[df["Month"] == 12]
print(december_data)
```

## Aggregations

Aggregate functions = Reduces a set of values into a single summary value, used to summarize and analyze data. Often used with the groupby() function

```python

# These aggregate function are for whole dataframe
# since we can't find mean of the non numeric columns we can specify inside of the mean function numeric only 
print(df.mean(numeric_only=True))

# we can also find the minimum and max values of the columns as well
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))

# wec an also coount the number of values in each column
print(df.count())


# now we can also use aggreagte functions with single columns
# mean of the meax temp column
print(df["Max Temp"].mean())

# groupby function is used to group the data by a specific column and then we can apply aggregate functions on the grouped data

# we want to find a mean max temperature for each month

group = df.groupby("Month")
mean_max_temp = group["Max Temp"].mean()
print(mean_max_temp)
```

## Data cleaning
The process of fixing/removing: incomplete, incorrect, or irrelevant data.
~75% of work done with Pandas is data cleaning

```python
# Let's say we want to drop a column cuz we don't need it anymore, we can use drop function
print(df.columns)
# 1. drop irrelevant column
df = df.drop(columns = ["Day"])
print(df.columns)

# 2. handle missing data
df = df.dropna(subset = ["Month"]) # this will drop all the rows that have missing values, any columns inside the subset list will be checked for missing values
# and if there is any missing value in those columns the whole row will be dropped

# instead of dropping we cna also fill the missing values with some value, for example 0
df = df.fillna(0) # this will fill all the missing values with 0

# we can also fill missing values with mean of the column which is called imputation
df = df.fillna(df.mean(numeric_only = True))
print(df)

```
